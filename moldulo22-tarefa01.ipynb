```python
import yfinance as yf
import pandas as pd
import plotly.express as px

ativos = ['^BVSP', 'USDBRL=X', 'VALE3.SA', 'PETR4.SA']
inicio = '2023-01-01'
fim = '2024-12-31'

df = yf.download(ativos, start=inicio, end=fim, auto_adjust=False)

# amplitude diária
amplitude = df['High'] - df['Low']

# média da amplitude por ativo
media_amplitude = amplitude.mean().sort_values(ascending=False)

# preparar para gráfico
amplitude_long = amplitude.reset_index().melt(
    id_vars='Date',
    var_name='Ativo',
    value_name='Amplitude'
)

fig = px.line(amplitude_long, x='Date', y='Amplitude', color='Ativo',
              title='Amplitude diária dos ativos')
fig.show()

print('Média da amplitude por ativo:')
print(media_amplitude)
```
