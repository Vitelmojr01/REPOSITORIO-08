1.Passo a passo do Bagging

Começa com o conjunto de dados original.
Faz várias amostragens com reposição (bootstrap), criando novos conjuntos com o mesmo tamanho do original, mas com linhas repetidas e outras de fora.
Treina um modelo em cada amostra. Esses modelos são os base learners.
Faz previsões com todos os modelos.
Junta os resultados na etapa de agregação:
classificação: voto da maioria;
regressão: média das previsões.




2.Bagging com minhas palavras
Bagging é uma técnica de ensemble que cria vários modelos a partir de amostras diferentes do mesmo dataset e depois combina as respostas para ter uma previsão mais estável e mais confiável. Ele ajuda principalmente a reduzir a variância e diminuir o risco de overfitting.


3.Opcional — código em Python

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

modelo = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=10,
    random_state=42
)

modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)

print("Acurácia:", accuracy_score(y_test, y_pred))
```
